# Analysing sales vs profit 

```python
category=df.groupby("Category")[["Sales","Profit"]].sum()
category
```

```python
df.plot('Sales','Profit',color='r',style='*')
plt.title('Sales vs Profit')  
plt.xlabel('Sales')  
plt.ylabel('Profit')  
plt.show()
```

From above graph we can conclude that on increse the slae from 0-15000 profit is also increasing but when sale >20000 then profit decreases.


# Analysising discount vs profit 

```python
df.plot('Discount','Profit',color='r',style='*')
plt.title('Discount vs Profit')  
plt.xlabel('Discount')  
plt.ylabel('Profit')  
plt.show()
```

From above graph we can conclude that on giving less discount more profit is done

# Analysising which ship mode makes more profit 

```python
sns.boxplot(x = 'Ship Mode', y = 'Profit', data = df,color='blue',saturation=1.5,width=0.8,linewidth=5)
```

From above graph we can conclude that Standard Class are giving the max profit and Same Day are giving the least.

# Analysising which segmant makes more profit 

```python
fig, ax=plt.subplots()
colors = {'Consumer':'red', 'Corporate':'green', 'Home Office':'yellow'}
ax.scatter(df['Sales'],df['Profit'], c=df['Segment'].apply(lambda x:colors[x]))
plt.show()
```

From above graph we can conclude that Corporates are giving the max profit and Home Office are giving the least.

# Analysising Category vs profit 

```python
df.plot('Category','Profit',color='r',style='*')
plt.title('Category vs Profit')  
plt.xlabel('Category')  
plt.ylabel('Profit')  
plt.show()
```

```python
temp_df = df.loc[(df['Category']=='Furniture') & (df['Discount']>=0.5)]
temp_df['Profit'].plot.hist(bins=40)
```

```python
temp_df = df.loc[(df['Category']=='Furniture') & (df['Discount']<0.5)]
temp_df['Profit'].plot.hist(bins=40)
```

```python
temp_df = df.loc[(df['Category']=='Technology') & (df['Discount']>=0.5)]
temp_df['Profit'].plot.hist(bins=40)
```

```python
temp_df = df.loc[(df['Category']=='Technology') & (df['Discount']<0.5)]
temp_df['Profit'].plot.hist(bins=40)
```

```python
temp_df = df.loc[(df['Category']=='Office Supplies') & (df['Discount']>=0.5)]
temp_df['Profit'].plot.hist(bins=40)
```

```python
temp_df = df.loc[(df['Category']=='Office Supplies') & (df['Discount']<0.5)]
temp_df['Profit'].plot.hist(bins=40)
```

From all the above graphs we can conclude that:

For all the categories (Office Supplies, Technology and Furniture):

    1. When discount is less than  50%, sales was going in profit.
    2. When discount is greater than or equal to 50%, sales was going in loss.
